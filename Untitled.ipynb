```python
# get assoicated_slides names
def get_slides_names(split_patient, df_id):
    WSI_In_split = []
    NoTumor = ['TNE0416', 'TNE0970', 'TNE0994', 'TNE1084', 'TNE1432']
    for p in split_patient:
        WSI_names = df_id[ df_id.iloc[:,1] == p].iloc[:,0].values
        for WSI in WSI_names:
            if WSI not in NoTumor:
                WSI_In_split.append(WSI)
    return WSI_In_split
```

```python
train_WSI = get_slides_names(train_sample, df_id)
val_WSI = get_slides_names(val, df_id)
```

```python
def tiles_list_split(split_list):
    list_tiles_sample = []
    for s in split_list:
   
        l_tiles = os.listdir(os.path.join('/data/gcs/lungNENomics/work/MathianE/Tiles_512_512_fullHENorm_from1802Dir', 
                   s, 'accept'))
        t_fpath_s = []
        for t in  l_tiles:
            t_fpath_s.append( os.path.join('/data/gcs/lungNENomics/work/MathianE/Tiles_512_512_fullHENorm_from1802Dir', 
                   s, 'accept', t))
        list_tiles_sample.append(t_fpath_s)
    return list_tiles_sample
```

```python
list_train_tiles =  tiles_list_split(train_WSI)
list_test_tiles =  tiles_list_split(test_WSI)
list_val_tiles =  tiles_list_split(val_WSI)
```

```python
# Get target
def diag_list_Tiles_split(WSI_split_list, dfImgInfo):
    # Typical 0; Atypical = 1 
    target = []
    for s in WSI_split_list:
        diag = dfImgInfo[dfImgInfo.iloc[:,0] == s]['Code Diag'].values[0]
        if diag == 'CA' or  diag.find('CA') != -1:
            target.append(1)
        else:
            target.append(0)
    return target
```

```python
target_train = diag_list_Tiles_split(train_WSI, dfImgInfo)
target_test = diag_list_Tiles_split(test_WSI, dfImgInfo)
target_val = diag_list_Tiles_split(val_WSI, dfImgInfo)
```
