```python
import os
import shutil 
import json
```

```python
with open('TrainNormalTumoral.json') as json_file:
    lib = json.load(json_file)
```

```python
main_output =  'TrainImg'
try:
    os.mkdir(main_output)
except:
    print('Mian folder already created')
for i in range(len(lib['Slides'])):
    sample = lib['Slides'][i]
    target = lib['Targets'][i]
    try:
        os.mkdir(os.path.join(main_output, str(target)))
    except:
        print('target folder {} already created'.format(target))
    for t in lib['Tiles'][i]:
        shutil.copy(t, os.path.join(main_output, str(target), t.split('/')[-1]))
        
```
